```python
# create a new dataset comparing response for all three models for the prompt - with three columns: prompt, accepted_response, rejected_response where accepted_response is the response with toxicity score < 0.5 and rejected_response is the response with toxicity score > 0.5

def get_accepted_rejected_data(df1, df2, df3):
    accepted_rejected_df = pd.DataFrame(columns=['Prompt','Accepted_response','Rejected_response'])
    for i in range(len(df1)):
        prompt = df1.iloc[i,0]
        response1 = df1.iloc[i,1]
        response2 = df2.iloc[i,1]
        response3 = df3.iloc[i,1]
        toxicity1 = df1.iloc[i,2]
        toxicity2 = df2.iloc[i,2]
        toxicity3 = df3.iloc[i,2]
        # get the minimum toxicity score of the three in accepted response
        min_toxicity = min(toxicity1,toxicity2,toxicity3)
        if toxicity1 == min_toxicity:
            accepted_response = response1
        elif toxicity2 == min_toxicity:
            accepted_response = response2
        elif toxicity3 == min_toxicity:
            accepted_response = response3
        # get the maximum toxicity score of the three in rejected response
        max_toxicity = max(toxicity1,toxicity2,toxicity3)
        if toxicity1 == max_toxicity:
            rejected_response = response1
        elif toxicity2 == max_toxicity:
            rejected_response = response2
        elif toxicity3 == max_toxicity:
            rejected_response = response3
        accepted_rejected_df = accepted_rejected_df.append({'Prompt':prompt,'Accepted_response':accepted_response,'Rejected_response':rejected_response},ignore_index=True)
    return accepted_rejected_df

accepted_rejected_df = get_accepted_rejected_data(model1_df, model2_df, model3_df)
accepted_rejected_df.head()

```

```python
import random
import pandas as pd
from operator import itemgetter
import torch
import warnings
warnings.filterwarnings('ignore')
from transformers import AutoModelForSequenceClassification,AutoTokenizer,TrainingArguments
from trl import RewardTrainer
```

```python
model_name = "distilroberta-base"
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=1)
tokenizer = AutoTokenizer.from_pretrained(model_name)
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token
    model.config.pad_token_id = model.config.eos_token_id
```

```python
def formatting_func(examples):
    kwargs = {"padding": "max_length", "truncation": True, "max_length": 512, "return_tensors": "pt"}
    prompt_plus_chosen_response = examples["Accepted_response"]
    prompt_plus_rejected_response = examples["Rejected_response"]
    tokens_chosen = tokenizer.encode_plus(prompt_plus_chosen_response, **kwargs)
    tokens_rejected = tokenizer.encode_plus(prompt_plus_rejected_response, **kwargs)
    return {
        "input_ids_chosen": tokens_chosen["input_ids"][0], "attention_mask_chosen": tokens_chosen["attention_mask"][0],
        "input_ids_rejected": tokens_rejected["input_ids"][0], "attention_mask_rejected": tokens_rejected["attention_mask"][0]
    }

from datasets import Dataset

# Assuming `formatted_dataset` is a DataFrame after applying `formatting_func`
dataset = Dataset.from_pandas(accepted_rejected_df)

formatted_dataset = dataset.map(formatting_func)

formatted_dataset = formatted_dataset.train_test_split()

from transformers import TrainerCallback



training_args = TrainingArguments(
    output_dir="drive/MyDrive/prompts/reward_distilbert_plot_model",
    per_device_train_batch_size=16,
    evaluation_strategy="epoch",
    logging_steps=1,
    num_train_epochs = 5,
    report_to=None,
)

trainer = RewardTrainer(
    model=model,
    args=training_args,
    tokenizer=tokenizer,
    train_dataset=formatted_dataset["train"],
    eval_dataset=formatted_dataset["test"],
)
trainer.train()
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model = model.to(device)
model
```

```python
trainer.save_model()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from torch.utils.data import DataLoader

test_dataloader = DataLoader(formatted_dataset["test"], batch_size=1)

model.eval()
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
predictions = []

with torch.no_grad():
    for batch in test_dataloader:
        inputs_accepted = tokenizer(batch['Accepted_response'], return_tensors='pt', padding='max_length', truncation=True, max_length=512).to(device)
        inputs_rejected = tokenizer(batch['Rejected_response'], return_tensors='pt', padding='max_length', truncation=True, max_length=512).to(device)

        outputs_accepted = model(**inputs_accepted)
        outputs_rejected = model(**inputs_rejected)

        score_accepted = torch.sigmoid(outputs_accepted.logits).squeeze().item()
        score_rejected = torch.sigmoid(outputs_rejected.logits).squeeze().item()

        predictions.append((score_accepted, score_rejected))

# Calculate the labels and scores
y_true = [1 for _ in predictions]  # 1 for each pair

# Create y_pred by comparing 'accepted' and 'rejected' scores for each pair
y_pred = [1 if accepted > rejected else 0 for accepted, rejected in predictions]

# Compute metrics
accuracy = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)




# Print the evaluation results
print(f"Accuracy: {accuracy}")
print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1 Score: {f1}")



```
